```python
# sprawdzę gdzie znajduje się kolumna `Garage.Yr.Blt`
data_updated.columns.get_loc("Garage.Yr.Blt")
```

```python
data_updated.head(3)
```

## Wartości porządkowe tzw. Ordinal Values

W data secie wystepują zmienne dla których przyjęte są zmienne porządkowe, tzw. ***ordinal values***, będące zmiennymi kategorycznymi. Są to wartości, które przedstawiają stopniowanie.
Dla przykładu, przedstawię unikalne wartości dla zmiennej `'External.Qual'`

```python
data_updated['Exter.Qual'].unique()
```

Zgodnie z dokumentacją wynika, że wartości dla zmiennej `'Exter.Qual'`oznaczają:
* `'Ex'` - Excellent
* `'Gd'` - Good
* `'TA'` - Average/Typical
* `'Fa'` - Fair
* `'Po'` - Poor

Jeżeli wykorzystamy sposób zamiany tych wartości kategorycznych na numeryczne jak w przypadku inncyh wartości kategorycznych model nie będzie wiedział, że te wartości odzwierciedlają stopniowanie.

Aby uniknąć tego problemu zamienię wartości porządkowe (kategoryczne) na numeryczne odzwierciedlające ranking. Czyli dla wartości `'NA'` przyjmiemy `0`, dla wartości `'Po'` - `1`, dla wartości `'Fa'` - `2` itd.

Zgodnie z dokumentacją wartości porządkowe przyjmują następujące zmienne:

`'Lot.Shape'`, `'Land.Slope'`, `'Electrical'`, `'Exter Qual'`, `'Exter Cond'`, `'Bsmt Qual'`, `'Bsmt Cond'`, `'Bsmt Exposure'`, `'BsmtFin Type 1'`, `'BsmtFinType 2'`, `'HeatingQC'`, `'Central Air'`, `'KitchenQual'`, `'FireplaceQu'`, `'Garage Finish'`, `'Garage Qual'`, `'Garage Cond'`, `'Pool QC'`, `'Fence'`

```python
map_1 = {'Reg':3,'IR1':2,'IR2':1,'IR3':0}
data_updated['Lot.Shape'] = data_updated['Lot.Shape'].map(map_1)

map_2 = {'Gtl':2,'Mod':1,'Sev':0}
data_updated['Land.Slope'] = data_updated['Land.Slope'].map(map_2)

map_3 = {'Ex':4,'Gd':3,'TA':2,'Fa':1,'Po':0}
data_updated['Exter.Qual'] = data_updated['Exter.Qual'].map(map_3)
data_updated['Exter.Cond'] = data_updated['Exter.Cond'].map(map_3)
data_updated['Heating.QC'] = data_updated['Heating.QC'].map(map_3)
data_updated['Kitchen.Qual'] = data_updated['Kitchen.Qual'].map(map_3)

map_4 = {'SBrkr':4, 'FuseA':3, 'FuseF':2,'FuseP':1,'Mix':0}
data_updated['Electrical'] = data_updated['Electrical'].map(map_4)

map_5 = {'Ex':5,'Gd':4,'TA':3,'Fa':2,'Po':1,'NA':0}
data_updated['Bsmt.Qual'] = data_updated['Bsmt.Qual'].map(map_5)
data_updated['Bsmt.Cond'] = data_updated['Bsmt.Cond'].map(map_5)

map_6 = {'Gd':4,'Av':3,'Mn':2,'No':1,'NA':0}
data_updated['Bsmt.Exposure'] = data_updated['Bsmt.Exposure'].map(map_6)

map_7 = {'GLQ':6,'ALQ':5,'BLQ':4,'Rec':3,'LwQ':2,'Unf':1,'NA':0}
data_updated['BsmtFin.Type.1'] = data_updated['BsmtFin.Type.1'].map(map_7)
data_updated['BsmtFin.Type.2'] = data_updated['BsmtFin.Type.2'].map(map_7)

map_8 = {'Y':1,'N':0}
data_updated['Central.Air'] = data_updated['Central.Air'].map(map_8)

map_9 = {'Ex':5,'Gd':4,'TA':3,'Fa':2,'Po':1,'NA':0}
data_updated['Fireplace.Qu'] = data_updated['Fireplace.Qu'].map(map_9)

map_10 = {'Fin':3,'RFn':2,'Unf':1,'NA':0}
data_updated['Garage.Finish'] = data_updated['Garage.Finish'].map(map_10)

map_11 = {'Ex':5,'Gd':4,'TA':3,'Fa':2,'Po':1,'NA':0}
data_updated['Garage.Qual'] = data_updated['Garage.Qual'].map(map_11)
data_updated['Garage.Cond'] = data_updated['Garage.Cond'].map(map_11)

map_12 = {'Ex':4,'Gd':3,'TA':2,'Fa':1,'NA':0}
data_updated['Pool.QC'] = data_updated['Pool.QC'].map(map_12)

map_13 = {'GdPrv':4,'MnPrv':3,'GdWo':2,'MnWw':1,'NA':0}
data_updated['Fence'] = data_updated['Fence'].map(map_13)
```

### Budowa Pipeline preprocessingu


Wykorzystamy DataframeSelector aby wybrać określone atrybuty z DataFrame:

```python
from sklearn.base import BaseEstimator, TransformerMixin

# A class to select numerical or categorical columns 
# since Scikit-Learn doesn't handle DataFrames yet
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names]
```

#### Budowa Pipeline dla atrybutow numerycnych
Zbudujmy pipeline dla atrybutów numerycznych:

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")

num_pipeline = Pipeline([
        ("select_numeric", DataFrameSelector([data_for_modeling_num])),
        ("imputer", SimpleImputer(strategy="median")),
    ])
```

```python
num_pipeline.fit_transform(X)
```

Będziemy także potrzebować imputera do kategorycznych kolumn napisowych (zwykły Imputer nie działa na tych kolumnach):

```python
# Inspired from stackoverflow.com/questions/25239958
class MostFrequentImputer(BaseEstimator, TransformerMixin):
    def fit(self, X, y=None):
        self.most_frequent_ = pd.Series([X[c].value_counts().index[0] for c in X],
                                        index=X.columns)
        return self
    def transform(self, X, y=None):
        return X.fillna(self.most_frequent_)
```

#### Budowa Pipeline dla atrybutow kategorycznych
Teraz możemy zbudować pipeline dla atrybutów kategorycznych:

```python
from sklearn.preprocessing import OneHotEncoder

cat_pipeline = Pipeline([
        ("select_cat", DataFrameSelector([data_for_modeling_cat])),
        ("imputer", MostFrequentImputer()),
        ("cat_encoder", OneHotEncoder(sparse=False)),
    ])
```
